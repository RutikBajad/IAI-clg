```python
import numpy as np
a = np.array([[2,-1],[-1,2]])
is_symmetric = True

for i in range(2):
    for j in range(2):
        if a[i][j] != a[j][i]:
            is_symmetric = False
            break
    if not is_symmetric:
        break

if is_symmetric:
    print("Matrix is symmetric...")
else:
    print("Matrix is not symmetric...")

z
```
